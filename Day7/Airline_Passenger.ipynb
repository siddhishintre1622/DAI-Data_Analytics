```python
airline_df = pd.read_csv('airline_passenger_satisfaction.csv')
ob_service = airline_df['On-board Service']

mean_ob_service = ob_service.mean()
median_ob_service = ob_service.median()
skewness_ob_service = ob_service.skew()

plt.figure(figsize=(8,6))
sns.histplot(ob_service, kde=True, color='skyblue', bins=30)
plt.axvline(mean_ob_service,color='red', linestyle='--', label=f'Mean: {mean_ob_service:.2f}')
plt.axvline(median_ob_service, color='black', linestyle='--',  label=f'Median: {median_ob_service:.2f}')
plt.title(f'Distribution of Distance\nSkewness: {skewness_ob_service:.2f}')
plt.xlabel('On-board Service')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
