```python
data = pd.read_csv('purchase_data.csv')
```

```python

# 5. Age Demographics
# o The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
#  Purchase Count
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
data['Age Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)
print("Purchase Count Age Wise :",data.groupby('Age Group')['Purchase ID'].count())
#  Average Purchase Price
print("Average Purchase price Age Wise :",data.groupby('Age Group')['Price'].mean())
#  Total Purchase Value
print("Total Purchase price Age Wise :",data.groupby('Age Group')['Price'].sum())
#  Average Purchase Total per Person by Age Group
print(data.groupby(['Age Group','SN'])['Price'].mean().dropna())
print('------------------------------------------------')



```
