```python
#load the dataset
df= pd.read_csv('Titanic-Dataset.csv')
```

```python
#Check Missing Values
missing_values = df.isnull().sum()
missing_values
```

```python
#Visualise Missing Values
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
#Analyze Categorical Features
sns.countplot(x='Sex', data=df)
plt.title('Count of Passengers by Sex')
plt.show()

sns.countplot(x='Embarked', data= df)
plt.title('Count of Passengers by Embarked Port')
plt.show()
```

```python
z_scores= stats.zscore(df['Fare'])
```

```python
z_scores
```

```python
sns.boxplot(x='Pclass', y= 'Fare', data=df)
```

```python
#Analyse Continuous Features
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Fare')
plt.show()
```

```python
#Analyse Survival Rate
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.show()

sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr= numeric_df.corr()
```

```python
corr
```

```python
#Correlation Matrix

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Pairplot of Selected Features')
plt.show()
```

```python
categorical_stats = {
    'Survived': df['Survived'].value_counts(normalize=True),
    'Sex': df['Sex'].value_counts(normalize=True),
    'Pclass': df['Pclass'].value_counts(normalize=True),
    'Embarked': df['Embarked'].value_counts(normalize=True),
}


print("Summary Statistics for Categorical Variables:")
for category, stats in categorical_stats.items():
    print(f"{category}:")
    print(stats)
```

```python
# Create a regression plot for Age vs. Fare
sns.regplot(x='Age', y='Fare', data=df)
plt.title('Regression Plot: Age vs. Fare')
plt.show()
```

```python
# Calculate Q1 and Q3
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['Fare'] < lower_bound) | (df['Fare'] > upper_bound)]
print(outliers)
```

```python
sns.boxplot(x='Fare', data=df)
plt.title('Box Plot of Fare to Identify Outliers')
plt.xlabel('Fare')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Scatter Plot of Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Create a scatter plot with a regression line
sns.regplot(x='Age', y='Fare', data=df, scatter_kws={'alpha':0.5})
plt.title('Scatter Plot of Age vs. Fare with Regression Line')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
