```python
#using matplotlib
plt.figure(figsize=(10,6))
plt.hist(df['Age'].dropna(),bins=30, color='blue', alpha=0.8)
plt.title("Distribution of Passenger Ages: ")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.grid(axis='y', alpha= 0.75)
plt.show()
```

```python
#using Seaborn

sns.set(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.set(style='darkgrid')
ax= sns.countplot(x='Survived', data= df, palette= 'pastel')
plt.title('Survival Counts')
plt.xlabel('Survived (0= No, 1= Yes)')
plt.ylabel('Number of Passengers')
#plt.xticks(ticks=[0,1], labels= ['No', 'Yes'])
plt.yticks(rotation=45)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
print(survival_counts)
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x= 'Fare',data= df, palette= 'pastel' )
plt.title('Distribution of Fare Prices')
plt.xlabel('Fare')
plt.show()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x= 'SibSp',data= df, palette= 'pastel' )
plt.title('Distribution of Number of Siblings/Spouses Abroad')
plt.xlabel('Number of Siblings/Spouses Abroad')
plt.ylabel('Number of Passenger')
plt.show()
```
