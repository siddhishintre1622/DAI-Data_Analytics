```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
ages = df.groupby('age').median().reset_index()
```

```python
ages.plot.line("age","chol")
```

```python
fig,ax =plt.subplots()
ax.plot(ages['age'],ages['chol'],ls=':',lw=1.7)
ax.set_xlabel('Age')
ax.set_ylabel('Cholestrol')
```

```python
fig,ax =plt.subplots()
ax.plot(ages['age'],ages['trestbps'],ls=':',lw=1.7)
ax.set_xlabel('Age')
ax.set_ylabel('Cholestrol')
```
