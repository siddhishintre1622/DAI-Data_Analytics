```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**4
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
```

```python
x = np.arange(0,10)
y = 2*x
print(x)
print(y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')

axes.plot(a,b,'r')
axes.set_xlabel('a')
axes.set_ylabel('b')

plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))
axes[0][0].plot(a,b)
axes[0][1].plot(b,a)
axes[1][0].plot(x,y)
axes[1][1].plot(y,x)

#fig.subplots_adjust(left= None, bottom= None, top=None, wspace=0.9, hspace = 0.1)
plt.show()
```
